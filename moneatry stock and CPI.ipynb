```python
import statsmodels.api as sm
```

```python
3. #REGRESSION ANALYSIS (Example Model)
# -------------------------------------------------
# ----1: Stock Return explained by Repo, Bank Rate, Reverse Repo ----
X = msc[['Bank rate', 'Repo rate', 'Reverse repo']]   # independent variables
y = msc['stock return']                              # dependent variable

X = sm.add_constant(X)  # Add intercept
model = sm.OLS(y, X).fit()

print("\n REGRESSION RESULT (Stock return ~ Bank, Repo, Reverse repo):")
print(model.summary())

```

```python
# 2: Stock Price explained by Inflation
X2 = msc[['combined inflation', 'Rural inflation', 'urban inflation']]  # independent variables
y2 = msc['Stock Price']    # dependent variable

X2 = sm.add_constant(X2)
model2 = sm.OLS(y2, X2).fit()

print("\n REGRESSION RESULT (Stock Price ~ CPI variables):")
print(model2.summary())
```

```python
# 2: Inflation explained by Rates
X2 = msc[['Bank rate', 'Repo rate', 'Reverse repo']]   #independent variables
y2 = msc['combined inflation']   #dependent variable

X2 = sm.add_constant(X2)
model2 = sm.OLS(y2, X2).fit()

print("\n REGRESSION RESULT (Inflation ~ Rates):")
print(model2.summary())
```
